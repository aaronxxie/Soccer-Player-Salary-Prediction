# 3. Data Cleaning

### Correcting I

#### Removing Duplicates

```python
# Check duplicate rows
train_df[train_df.duplicated(keep = False)==True].sort_values(by = ['ID']).head()
```

```python
# Remove all duplicates and check the results
train_df = train_df.drop_duplicates()
test_df = test_df.drop_duplicates()
train_df.duplicated(keep = False).sum()
```

#### Handling Missing Values

```python
# Checking columns with missing values
print('Train Dataframe:\n', train_df.isnull().sum()[train_df.isnull().sum() > 0])
print('Test Dataframe:\n', test_df.isnull().sum()[test_df.isnull().sum()> 0])
```

* 'ClubNumber' and 'NationalNumber' columns are already excluded because they have no impact on target variables. 
* Since NationalPosition has too many null values, it should be excluded, too. 

* For 'ClubPosition', I will fill it with a new category "NP" (No Position).
* For 'ContractUntil', I will fill it with median. 

```python
# Checking players with 0 salary
train_df[train_df.WageEUR == 0].shape
```

There are 71 players having no salary. I need to exclude these players because they are those who currently do not have a contract yet. They deserve more than just 0 salary.

```python
# Dropping columns
column_todrop = ['ID', 'Name', 'FullName', 'PhotoUrl', 'ClubNumber', 'NationalNumber', 'NationalPosition']
train_df = train_df.drop(column_todrop, axis=1)
test_df = test_df.drop(column_todrop, axis=1)
# Dropping rows with 0 WageEUR
train_df = train_df.drop(train_df[train_df.WageEUR == 0].index)
test_df = test_df.drop(test_df[test_df.WageEUR == 0].index)
cleaner = [train_df, test_df]
```

### Creating

#### Feature Engineering

```python
for dataset in cleaner:
    # Transform 'ClubJoined' and 'ContractUntil'
    dataset['ClubJoined'] = 2022 - dataset['ClubJoined']
    dataset['ContractUntil'] = dataset['ContractUntil'] - 2022
    # Creating the feature "PositionNum" by counting the number of "," in columns
    dataset['PositionNum'] = dataset['Positions'].str.count(',')+1
```

```python
train_df[['ClubJoined','ContractUntil','PositionNum']].head()
```

### Completing

```python
# Filling Null Values
for dataset in cleaner:
    dataset['ClubPosition'] = dataset['ClubPosition'].fillna('NP')
    dataset['ContractUntil'] = dataset['ContractUntil'].fillna(dataset['ContractUntil'].median())
```

```python
# Checking Null Values again
train_df.isnull().sum()[train_df.isnull().sum() > 0]
```

### Correlation Analysis

```python
# Creating categorical variable list
categorical_variables = []
for col in list(train_df.columns):
    if train_df[col].dtypes == object or train_df[col].dtypes == bool: 
        categorical_variables += [col]
        
# Creating numerical variable list
numerical_variables = []
for col in list(train_df.columns):
    if train_df[col].dtypes == np.int64 or train_df[col].dtypes == np.float64: # note: don't just write 'int64'
        numerical_variables += [col]
numerical_variables.remove('WageEUR')
#numerical_variables
```

Some variables in "numerical_variables" look like discrete variables, but they are actually ordinal variables. The difference is there is a uniform degree of difference within discrete quantitative variables.

```python
# Create ordinal variable list
ordinal_variables = ['IntReputation','WeakFoot','SkillMoves','PositionNum']
# Remove ordinal variable from numerical variables
numerical_variables = list(set(numerical_variables) - set(ordinal_variables))
# Add ordinal variables to categorical variables
categorical_variables = categorical_variables + ordinal_variables
ordinal_variables = ordinal_variables + ['AttackingWorkRate','DefensiveWorkRate']
```

#### Target (Numerial) and Categorical Variables

```python
# Pivoting features to check mean differences
print("\nMean Differences")
print("-"*60)
for elem in categorical_variables:
    print(train_df[[elem, 'WageEUR']].groupby([elem], as_index=False).mean().sort_values(by='WageEUR', ascending=False))

# Pivoting features to check median differences
#print("\nMedian Differences")
#print("-"*60)
#for elem in categorical_variables:
    #print(train_df[[elem, 'WageEUR']].groupby([elem], as_index=False).median().sort_values(by='WageEUR', ascending=False))
```

Some Inferences:
* All categorical variables seem to correlate with target variable.
* The new feature **"PositionNum"** from feature engineering indicates that the more positions a player can play, the higher salary can the player earn.
* The Central Forward (CF) position has the highest average salary.
* The Manchester City club has has the highest average salary.
